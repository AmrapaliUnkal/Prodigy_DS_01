# Data Cleaning

```python
df.isna().sum()
```

```python
df=df.fillna(method='ffill')
```

```python
df=df.drop('2023',axis=1)
```

```python
years=df.describe().columns
```

# Plotting a bar chart

```python
plt.figure(figsize=(30,30))
plt.bar(years,df[years].sum())
plt.xlabel("Year",size=40)
plt.xticks(rotation=45)
plt.ylabel("Population (Billion)",size=40)
plt.title("Total Population Per Year",size=50)
plt.show()
```

```python
plt.figure(figsize=(30,30))
plt.barh(years,df[years].sum())
plt.xticks(rotation=45)
plt.xlabel("Population (Billion)",size=40)
plt.ylabel("Year",size=40)
plt.title("Total Population Per Year",size=50)
plt.show()
```
